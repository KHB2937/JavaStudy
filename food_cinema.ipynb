```python
import pandas as pd

# CSV 파일 경로
file_path = '/content/drive/MyDrive/Colab Notebooks/fulldata_07_24_04_P_일반음식점.csv'
```

```python
# CSV 파일 읽기
data = pd.read_csv(file_path, encoding='cp949', low_memory=False)

# 데이터 확인
# data
```

```python
data = data[data['영업상태명'] != '폐업']
```

```python
data
```

```python
# 삭제할 열 목록
columns_to_drop = ['번호', '개방서비스아이디', '개방자치단체코드', '관리번호', '인허가취소일자',
                   '공장생산직직원수', '건물소유구분명', '보증액', '월세액', '전통업소지정번호',
                   '홈페이지', 'Unnamed: 47', '폐업일자', '휴업시작일자', '휴업종료일자', '재개업일자', '영업장주변구분명',
                   '개방서비스명', '인허가일자', '영업상태구분코드', '상세영업상태코드', '상세영업상태명',
                   '데이터갱신일자', '위생업태명', '데이터갱신구분', '도로명우편번호', '도로명전체주소',
                   '여성종사자수', '등급구분명', '급수시설구분명', '총직원수','본사직원수', '공장사무직직원수',
                   '공장판매직직원수', '다중이용업소여부', '시설총규모', '전통업소주된음식', '남성종사자수', '소재지면적', '소재지우편번호', '최종수정시점']
data = data.drop(columns=columns_to_drop)
data
```

```python
data
```

```python
# 데이터프레임 인덱스 재배열
data = data.reset_index(drop=True)
# data
```

```python
from pyproj import Proj, transform
import numpy as np
import pandas as pd

# 좌표 변환을 위한 좌표 체계 설정
proj_bessel_tm = Proj(init='epsg:5174')  # Bessel 중부원점 TM 좌표체계
proj_wgs84 = Proj(init='epsg:4326')  # WGS84 좌표체계

# 좌표 변환 함수
def convert_coordinates(x, y):
    x_, y_ = transform(proj_bessel_tm, proj_wgs84, x, y)
    return x_, y_

# 배치 단위 크기 설정
batch_size = 500

# 데이터프레임을 배치 단위로 처리
for i in range(0, len(data), batch_size):
    batch_data = data[i:i+batch_size].copy()  # 배치 데이터 선택
    batch_data['lon'], batch_data['lat'] = zip(*batch_data.apply(lambda row: convert_coordinates(row['좌표정보(x)'], row['좌표정보(y)']), axis=1))
    data[i:i+batch_size] = batch_data  # 변환된 좌표로 업데이트

    # 현재 처리된 데이터 개수 출력
    current_count = min(i + batch_size, len(data))
    print(f"현재 처리된 데이터 개수: {current_count}/{len(data)}")

# 변환 결과 출력
data
```

```python
# 수정된 데이터프레임을 CSV 파일로 저장
output_file_path = './전국음식점(수정본).csv'
data.to_csv(output_file_path, index=False)

print("CSV 파일이 저장되었습니다.")
```
